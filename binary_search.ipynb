__binary search__

두 리스트를 받아서 한 리스트의 원소가 다른 리스트에 있는지 확인하는 문제  

시간 제한 : 1초  
입력 : 자연수 1쌍, 각 자연수 값만큼 리스트 1쌍, 리스트의 원소들은 -2^31 ~ 2^31의 범위의 정수  

## 분석
원소의 범위가 커서 '그' 탐색 X  
첫 리스트는 정렬해야 해서 (n log(n)) --> 내장 정렬 알고리즘  
이진 분류를 통해 값 찾기 (log(n) x len(list))  

_16*100000 + 16*100000 = 3200000 --> 대충 0.32초?_

## 결과
실제로는 0.516초, 0.676초 나왔다.  
재귀 함수를 사용하는 것이 더 오래걸렸다.  

문제에서 요구하는 출력이 입력 리스트의 개수만큼 잦아서  
문자열로 저장한 뒤 한번에 출력하는 것이 성능 향상에 유의미했다.  

예상과 크게 차이가 났다.  
원인은 잘 모르겠지만 현재 생각으로는  
단순히 입력 크기로 복잡도를 계산하기 보다는  
실제 연산에 필요한 양이  
 - 비교문 --> 평균 2.5
 - 중간값 계산 --> 1
 - 기타 연산 --> 1
이므로 다시 계산해보면  

_16 + log(4.5 * 1000000) + 1 = 0.36초 정도_  
... 잘 모르겟다 ㅋ

```python
import sys
sys.stdin = open('1920_input.txt')
input = sys.stdin.readline
```

```python
# 1920-1
N = int(input())
N_arr = set(map(int, input().split()))

M = int(input())
M_arr = list(map(int, input().split()))

r = ""
for ele in M_arr:
    if ele in N_arr:
        r += '1\n'
    else:
        r += '0\n'
print(r)
```

```python
# 1920-2
def binary_search(
    arr: list, #sorted list
    target: int,
    start: int,
    end: int):
    if start > end:
        return None
    
    mid = (start+end)//2
    if arr[mid] == target:
        return mid+1
    elif arr[mid] > target:
        return binary_search(arr, target, start, mid-1)
    else:
        return binary_search(arr, target, mid+1, end)

N = int(input())
N_arr = sorted(list(map(int, input().split())))

M = int(input())
M_arr = list(map(int, input().split()))

r = ""
for ele in M_arr:
    if binary_search(N_arr, ele, 0, N-1):
        r += "1\n"
    else:
        r += "0\n"
print(r)
```

```python
# 1920-3
# difference of implementing binary search
def binary_search(
    arr: list, #sorted list
    target: int):    
    start, end = 0, len(arr)-1
    
    while(True):
        if start > end:
            return None

        mid = (start+end)//2
        if target == arr[mid]:
            return mid+1
        elif target > arr[mid]:
            start = mid+1
        else:
            end = mid-1


N = int(input())
N_arr = sorted(list(map(int, input().split())))

M = int(input())
M_arr = list(map(int, input().split()))

r = ""
for ele in M_arr:
    if binary_search(N_arr, ele):
        r += "1\n"
    else:
        r += "0\n"
print(r)
```
